```python
from qiskit import *
from qiskit.visualization import plot_bloch_vector
%matplotlib inline
```

```python
plot_bloch_vector([0,0,1], title="Spin-Up")
```

```python
plot_bloch_vector([0,0,-1], title="Spin-Down")
```

```python
plot_bloch_vector([0,1,0], title="Spin-Right")
```

```python
circuit = QuantumCircuit(1,1)
```

```python
%matplotlib inline
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
#Import Aer
from qiskit import Aer

#Run the quantum circuit on a statevector backend
backend = Aer.get_backend('statevector_simulator')
```

```python
#Create a quantum circuit for execution
job = execute(circuit, backend)
```

```python
result = job.result()
```

```python
outputstate = result.get_statevector(circuit, decimals=3)
print(outputstate)
```
